```python
import matplotlib.pyplot as plt
```

```python
x = range(100)
y = [value**2 for value in x]
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
x_1 = range(100)
y_1 = [np.cos(value) for value in x]

x_2 = range(100)
y_2 = [np.sin(value) for value in x]

plt.plot(x_1,y_1)
plt.plot(x_2,y_2)
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10,5)

ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(x_1, y_1, c='b')
ax_2.plot(x_2, y_2, c='g')
plt.show()
```

```python
x_1 = np.arange(100)
y_1 = np.arange(100) 

x_2 = np.arange(100)
y_2 = np.arange(100) + 100

plt.plot(x_1, y_1, color='maroon')
plt.plot(x_2, y_2, c = 'r')
plt.show()
```

```python
plt.plot(x_1, y_1, color='maroon', linestyle='dashed')
plt.plot(x_2, y_2, c = 'r', ls = 'dotted')
plt.show()
```

```python
plt.plot(x_1, y_1, color='maroon', linestyle='dashed')
plt.plot(x_2, y_2, c = 'r', ls = 'dotted')
plt.title('two lines')
plt.show()
```

```python
plt.plot(x_1, y_1, color='maroon', linestyle='dashed', 
         label='y=x')
plt.plot(x_2, y_2, c = 'r', ls = 'dotted', label = 'y=x+100')
plt.title('two lines')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.plot(x_1, y_1, color='maroon', linestyle='dashed', 
         label='y=x')
plt.plot(x_2, y_2, c = 'r', ls = 'dotted', label = 'y=x+100')
plt.title('two lines')
plt.legend(loc='lower right')
plt.grid(lw = 0.4, ls = 'dashed', c='.90')
plt.xlim(-100, 100)
plt.ylim(-200, 200)
plt.show()
```

```python
data1 = np.random.rand(512, 2)
data2 = np.random.rand(512, 2)

plt.scatter(data1[:,0], data1[:,1], c='r', marker='x')
plt.scatter(data2[:,0], data2[:,1], s=0.5,c='g')
plt.show()
```

```python
n = 50

x = np.random.rand(n)
y = np.random.rand(n)

colors = np.random.rand(n)
area = (np.random.rand(n)*1000)

plt.scatter(x,y, s=area, c=colors)
plt.show()
```

```python
data = [[5,25,50,20],
       [4,23,51,17],
       [6,22,52,19]]

X = np.arange(0,8,2)
plt.bar(X, data[0], color='b', width=0.5)
plt.bar(X+0.5, data[1], color='g', width=0.5)
plt.bar(X+1, data[2], color='r', width=0.5)
plt.xticks(X+0.5, ('q','b','c','d'))

plt.show()
```

```python
color_lst = ['b','g','r']
data_label = ['a','b','c']

x = np.arange(np.array(data).shape[1])

plt.bar(x, data[0], bottom = np.sum(data[:0], axis=0),
       color = color_lst[0], label = data_label[0])
plt.bar(x, data[1], bottom = np.sum(data[:1], axis=0),
       color = color_lst[1], label = data_label[1])
plt.bar(x, data[2], bottom = np.sum(data[:2], axis=0),
       color = color_lst[2], label = data_label[2])

plt.legend()
plt.show()
```
