# Series

```python
from pandas import Series, DataFrame
import pandas as pd
```

```python
example_obj = Series()
```

```python
list_data = [1,2,3,4,5]
example_obj = Series(data=list_data)
example_obj
```

```python
list_data = [1,2,3,4,5]
list_name = ['a', 'b', 'c', 'd', 'e']
example_obj = Series(data=list_data, index=list_name)
example_obj
```

```python
import numpy as np
dict_data = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5}
example_obj = Series(dict_data, dtype=np.float32, name='example_data')
example_obj
```

```python
example_obj['a']
```

```python
example_obj['a'] = 3.2
example_obj
```

```python
example_obj.values
```
